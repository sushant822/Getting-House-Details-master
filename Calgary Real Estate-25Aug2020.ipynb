```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
import time
print("Libraries imported!")
```

```python
url = 'https://www.remax.ca/ab/calgary-real-estate'
page = requests.get(url)
soup = BeautifulSoup(page.text, "html.parser")
```

```python
house_address = []
house_details = []
house_details_more=[]
url_house = []
year_built = []
pc = []
lat = []
long = []

base_url = 'https://www.remax.ca/ab/calgary-real-estate?page='
urls = [base_url + str(x) for x in range(1,1)]

for url in urls:
    
    #browser.visit(url)
    time.sleep(5)

    # Parse HTML with Beautiful Soup
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    try:
        addresses = soup.find_all('div', class_='left-content flex-one')
        for address in addresses:
            house_address.append(address.text)
    except:
        house_address.append('None')
        
    try:
        details = soup.find_all('div', class_='property-details')
        for detail in details:
            house_details.append(detail.text)
    except:
        house_details.append('None')
        
    try:
        for detail in soup.find_all('a'):
            house_details_more.append(detail.get('href'))
    except:
        house_details_more.append('None')
        
    for i in house_details_more:
        if '/ab/calgary-real-estate/' in str(i):
            link = 'https://www.remax.ca' + str(i)
            url_house.append(link)
    time.sleep(5)      
    for i in url_house:
        time.sleep(5)
        page = requests.get(i)
        soup = BeautifulSoup(page.text, "html.parser")
        house_details = []
        try:
            addresses = soup.find_all('section', class_='property-details')
            for address in addresses:
                house_details.append(address.text)
        except:
            house_details.append('None')
        string = ', '.join(house_details)

        s_year = "Year Built"
        position_year = string.find(s_year)
        year_built_temp = string[(position_year+12):(position_year+16)]
        year_built.append(year_built_temp)


        try:
            cor_all = soup.find_all('script')[4]
            for cor in cor_all:
                house_details.append(cor_all.text)
        except:
            house_details.append('None')
        string = ', '.join(house_details)

        string = ', '.join(cor_all)
        lat_t = "latitude"
        position_lat = string.find(lat_t)
        latd = string[(position_lat+11):(position_lat+20)]
        lat.append(latd)

        long_t = "longitude"
        position_long = string.find(long_t)
        longd = string[(position_long+12):(position_long+23)]
        long.append(longd)

        pct = "postalCode"
        position_pc = string.find(pct)
        pcd = string[(position_pc+14):(position_pc+20)]
        pc.append(pcd)
        time.sleep(5)
```
